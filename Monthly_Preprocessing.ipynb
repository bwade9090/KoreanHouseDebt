## BTC 거래대금

### Upbit

```python
df_upbit = pd.read_csv('data/krw_btc_monthly_upbit.csv')
```

```python
df_upbit = df_upbit[::-1]
```

```python
df_upbit.index = pd.date_range(start='2017-09-30', end='2022-01-31', freq='M')
```

```python
df_upbit.candle_acc_trade_price.plot(figsize=(15, 8))
```

### Bithumb

```python
df_bithumb = pd.read_csv('data/bithumb_btc_daily.csv')
```

```python
df_bithumb['krw_transaction'] = df_bithumb['c'] * df_bithumb['v']
```

```python
df_bithumb['yyyymm'] = df_bithumb.time.apply(lambda x: date.fromisoformat(x).strftime('%Y%m'))
```

```python
df_bithumb.groupby('yyyymm').sum().krw_transaction.plot(figsize=(15, 8))
```

```python
df_bithumb_monthly = df_bithumb.groupby('yyyymm').sum()
```

```python
df_bithumb_monthly.index
```

```python
for y in range(2014, 2022):
  for m in range(1, 13):
    if '{}{:02}'.format(y, m) not in df_bithumb_monthly.index:
      print('{}{:02}'.format(y, m))
```

```python
df_bithumb_monthly = df_bithumb_monthly[23:]
```

```python
df_bithumb_monthly.index = pd.date_range(start='2016-01-31', end='2022-02-28', freq='M')
```

### Coinone

```python
df_coinone = pd.read_csv('data/coinone_btc_daily.csv')
```

```python
df_coinone.dtypes
```

```python
df_coinone['krw_transaction'] = df_coinone['Close'] * df_coinone['Volume']
```

```python
df_coinone['yyyymm'] = df_coinone.time.apply(lambda x: date.fromisoformat(x).strftime('%Y%m'))
```

```python
df_coinone.groupby('yyyymm').sum().krw_transaction.plot(figsize=(15, 8))
```

```python
df_coinone_monthly = df_coinone.groupby('yyyymm').sum()
```

```python
df_coinone_monthly = df_coinone_monthly[17:]
```

```python
df_coinone_monthly.index = pd.date_range(start='2016-01-31', end='2022-02-28', freq='M')
```

### Sum

```python
df = pd.DataFrame(index=pd.date_range(start='2016-01-31', end='2022-02-28', freq='M'))
```

```python
df['upbit'] = df_upbit['candle_acc_trade_price']
df['bithumb'] = df_bithumb_monthly['krw_transaction']
df['coinone'] = df_coinone_monthly['krw_transaction']
```

```python
df.plot.area(figsize=(15, 8))
```

```python
df = df.fillna(0)
```

```python
df['btc_krw_trans'] = df['upbit'] + df['bithumb'] + df['coinone']
```

```python
crypto_df['btc_krw_transaction'] = df['btc_krw_trans']
```
