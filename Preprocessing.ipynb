# 주요 거시지표

## 기준분기

```python
base_quarter = pd.read_csv('data/2110059_20220113180614.csv')['기준분기'].unique()
```

```python
base_quarter
```

```python
all_df = pd.DataFrame(index=base_quarter)
```

## m2 말잔

```python
df = pd.read_excel('data/M2_상품별구성내역_말잔_원계열__20220117172315.xlsx')
```

```python
df = df.iloc[48:]
```

```python
all_df['m2'] = list(df['M2 (말잔,원계열)'])
```

## 본원통화 말잔

```python
df = pd.read_excel('data/본원통화_구성내역_말잔_원계열__20220117172516.xlsx')
```

```python
df = df.iloc[32:]
```

```python
all_df['main_currency'] = list(df['본원통화(말잔,원계열)'])
```

## 대출금리 - 가계대출

```python
df = pd.read_excel('data/신규취급액_기준_20220117174047.xlsx')
```

```python
df = df.iloc[48:]
```

```python
all_df['loan_ir'] = list(df['가계대출 (연리%)'])
```

## 주택매매가격지수

```python
df = pd.read_csv('data/월간_매매가격지수_종합.csv')
```

```python
df = df.iloc[2::3]
```

```python
df['기준시점'] = base_quarter
```

```python
df = df.set_index('기준시점')
```

```python
df = df.filter(['지수']).rename(columns={'지수':'house_price_index'})
```

```python
all_df = all_df.join([df])
```

## 주택평균거래량

```python
df = pd.read_csv('data/주택거래현황.csv')
```

```python
df = df.iloc[48:-2]
```

```python
df['분기'] = sorted(list(base_quarter) * 3)
```

```python
all_df['house_transaction_volumne'] = list(df.groupby('분기').sum()['합계'])
```

## 기준금리

```python
df = pd.read_csv('data/기준금리.csv')
```

```python
all_df['base_rate'] = list(df['원자료'])
```

## 코스피 코스닥

```python
df = pd.read_csv('data/코스피코스닥.csv')
```

```python
df = df.iloc[2::3]
```

```python
import locale
locale.setlocale(locale.LC_ALL, 'en_US.UTF-8')
```

```python
all_df['kospi'] = list(df['코스피지수'].apply(locale.atof))
all_df['kosdaq'] = list(df['코스닥지수'].apply(locale.atof))
```

## 가계예금

```python
df = pd.read_csv('data/가계예금.csv')
```

```python
import locale
locale.setlocale(locale.LC_ALL, 'en_US.UTF-8')
```

```python
all_df['house_savings'] = list(df['원자료'].apply(locale.atof))
```

## 소비자물가지수

```python
df = pd.read_csv('data/소비자물가지수.csv')
```

```python
all_df['cpi'] = list(df['전국'])
```

## export

```python
all_df.dtypes
```

```python
all_df.to_csv('data/all_macro_data.csv')
```
